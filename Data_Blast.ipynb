```python
# Liste pour stocker les DataFrames filtrés
filtered_dfs = []

for genome_file in genome_files:
    # Charger le fichier
    df = pd.read_csv(genome_file, delimiter="\t", skiprows=1, header=None)
    
    # Appliquer les noms des colonnes
    df.columns = noms_colonnes
    
    # Grouper les TE par chromosome
    grouped = df.groupby('Chr')  
    for chr_name, group in grouped:
        # Trier les TE par leur position de début (TE.start)
        group = group.sort_values(by='TE.start').reset_index(drop=True)

        # Liste pour stocker les indices des TE à conserver
        to_keep = []

        # Variable pour garder la fin du dernier TE conservé
        last_end = -float('inf')

        # Parcourir chaque ligne du DataFrame et garder les TE qui n'ont pas de TE proches
        for i, row in group.iterrows():
            if row['TE.start'] - last_end > 500:
                to_keep.append(i)
                last_end = row['TE.end']

        # Filtrer le DataFrame pour conserver uniquement les TE sélectionnés
        filtered_group = group.loc[to_keep]
        filtered_dfs.append(filtered_group)

# Combiner les DataFrames filtrés pour chaque génome
filtered_te_df = pd.concat(filtered_dfs).reset_index(drop=True)

# Sauvegarder le DataFrame filtré dans un nouveau fichier CSV
filtered_te_df.to_csv('filtered_te_data.csv', index=False)

print(f'Original number of TEs: {sum([len(pd.read_csv(genome_file, delimiter="\t", skiprows=1)) for genome_file in genome_files])}')
print(f'Filtered number of TEs: {len(filtered_te_df)}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
import warnings

df_cleaned['TE.superfamily_binary'] = df_cleaned['TE.superfamily'].map(lambda x: 1 if x == 'Helitron' else 0)


# Ignorer les avertissements pour une meilleure lisibilité
warnings.filterwarnings("ignore", category=UserWarning, module="matplotlib")

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(numeric_df, df_cleaned['TE.superfamily_binary'], test_size=0.2, random_state=42)

# Définir le modèle RandomForestClassifier
model = RandomForestClassifier(max_depth=None, min_samples_leaf=1, min_samples_split=2, n_estimators=300)


# Effectuer la validation croisée avec 5 folds
cv_results = cross_validate(model, X_train, y_train, cv=5, return_train_score=True, return_estimator=True)

# Récupérer les estimateurs de modèle pour chaque fold
estimators = cv_results['estimator']

# Tracer la matrice de confusion pour chaque fold
plt.figure(figsize=(15, 10))
for i, estimator in enumerate(estimators):
    try:
        # Prédictions sur les données de test
        predictions = estimator.predict(X_test)

        # Calculer la matrice de confusion
        conf_matrix = confusion_matrix(y_test, predictions)

        # Tracer la matrice de confusion
        plt.subplot(2, 3, i+1)
        sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt="d")
        plt.title(f"Confusion Matrix - Fold {i+1} Helitron")
        plt.xlabel("Predicted labels")
        plt.ylabel("True labels")
    except Exception as e:
        print(f"Fold {i+1} a échoué avec l'erreur : {e}")

plt.tight_layout()
plt.show()
```
