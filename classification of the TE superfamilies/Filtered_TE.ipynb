```python
import pandas as pd

# Noms des fichiers des 89 génomes
genome_files = [
    "C24.summary.bed", "Lag1-2.summary.bed", "Lag1-7.summary.bed", "Krazo_1.summary.bed",
    "OOE3-1.summary.bed", "Castelfeld_4_210.summary.bed", "Castelfeld_4_211.summary.bed", "CATS6.summary.bed",
    "Boo2-3.summary.bed", "Dolna_1_40.summary.bed", "IP-Fun0.summary.bed", "IP-Gra0.summary.bed",
    "IP_Her12.summary.bed", "IP_Vdm.summary.bed", "Esp1-11.summary.bed", "IP_Nog17.summary.bed",
    "Kru3.summary.bed", "Lov5.summary.bed", "NFA10.summary.bed", "Qar8a.summary.bed",
    "Sri3.summary.bed", "Sij1.summary.bed", "Yeg7.summary.bed", "T930.summary.bed",
    "Dolna1-10.summary.bed", "En_D.summary.bed", "Tac_0.summary.bed", "UKSW06_257.summary.bed",
    "627RMX-1MN5.summary.bed", "D-UKSW06-491.summary.bed", "KER5.summary.bed", "NYL7.summary.bed",
    "TOM-01.summary.bed", "UKSW06-541.summary.bed", "Di-G.summary.bed", "IP-Alm-0.summary.bed",
    "IP-Moa-0.summary.bed", "IP-Pun0.summary.bed", "Ler_0.summary.bed", "LI-WP-039.summary.bed",
    "MIA-1.summary.bed", "Eden5.summary.bed", "IP-Per0.summary.bed", "KNO2-54.summary.bed",
    "Nyl13.summary.bed", "T890.summary.bed", "Angso-74-430.summary.bed", "Ba5-1.summary.bed",
    "Cvi-0.summary.bed", "Eden1.summary.bed", "IP-Vaz0.summary.bed", "IP-Vis0.summary.bed",
    "LI-EF-011.summary.bed", "OOE2-1.summary.bed", "Taa14.summary.bed", "UKID93.summary.bed",
    "UKSW06-240.summary.bed", "Zu-0.summary.bed", "Bach2-1.summary.bed", "Bur-0.summary.bed",
    "Kas-2.summary.bed", "Lag2-4.summary.bed", "MNF-Pin40.summary.bed", "Fri1.summary.bed",
    "Samm.summary.bed", "TV30.summary.bed", "Lov1.summary.bed", "Anzo.summary.bed",
    "Eds9.summary.bed", "IP-Pal0.summary.bed", "IP-Pan0.summary.bed", "Lerik1-7.summary.bed",
    "Lerik2-7.summary.bed", "IP-Cat0.summary.bed", "IP-Cem0.summary.bed", "IP-Tol7.summary.bed",
    "Kas1.summary.bed", "Yeg8.summary.bed", "OOE23.summary.bed", "Etna2.summary.bed",
    "MIA-5.summary.bed", "Ta_0.summary.bed", "Tsu0.summary.bed", "IP_Hum2.summary.bed",
    "Sf_2.summary.bed", "T610.summary.bed", "col_g2-1_dd.summary.bed", "Ct_1.summary.bed",
    "T1160.summary.bed"
]

# Noms des colonnes pour tous les génomes
noms_colonnes = [
    "Chr", "TE.start", "TE.end", "TE.strand", "TE.id", "TE.family", "TE.superfamily",
    "up.Gene1.start", "up.Gene1.end", "up.Gene1.id", "up.Gene1.strand", "up.Gene1.distance",
    "up.Gene2.start", "up.Gene2.end", "up.Gene2.id", "up.Gene2.strand", "up.Gene2.distance",
    "down.Gene1.start", "down.Gene1.end", "down.Gene1.id", "down.Gene1.strand", "down.Gene1.distance",
    "down.Gene2.start", "down.Gene2.end", "down.Gene2.id", "down.Gene2.strand", "down.Gene2.distance",
    "up.Meth.CG.0_100", "up.Meth.CHG.0_100", "up.Meth.CHH.0_100", "up.Meth.CG.100_200",
    "up.Meth.CHG.100_200", "up.Meth.CHH.100_200", "up.Meth.CG.200_300", "up.Meth.CHG.200_300",
    "up.Meth.CHH.200_300", "down.Meth.CG.0_100", "down.Meth.CHG.0_100", "down.Meth.CHH.0_100",
    "down.Meth.CG.100_200", "down.Meth.CHG.100_200", "down.Meth.CHH.100_200", "down.Meth.CG.200_300",
    "down.Meth.CHG.200_300", "down.Meth.CHH.200_300", "TE.Meth.CG", "TE.Meth.CHG", "TE.Meth.CHH",
    "up.context.CG.0_100", "up.context.CHG.0_100", "up.context.CHH.0_100", "up.context.CG.100_200",
    "up.context.CHG.100_200", "up.context.CHH.100_200", "up.context.CG.200_300", "up.context.CHG.200_300",
    "up.context.CHH.200_300", "down.context.CG.0_100", "down.context.CHG.0_100", "down.context.CHH.0_100",
    "down.context.CG.100_200", "down.context.CHG.100_200", "down.context.CHH.100_200", "down.context.CG.200_300",
    "down.context.CHG.200_300", "down.context.CHH.200_300", "TE.context.CG", "TE.context.CHG", "TE.context.CHH"
]

# Liste pour stocker les DataFrames filtrés
filtered_dfs = []

for genome_file in genome_files:
    # Charger le fichier
    df = pd.read_csv(genome_file, delimiter="\t", skiprows=1, header=None)
    
    # Appliquer les noms des colonnes
    df.columns = noms_colonnes
        
    # Grouper les TE par chromosome
    grouped = df.groupby('Chr')  
    for chr_name, group in grouped:
        # Trier les TE par leur position de début (TE.start)
        group = group.sort_values(by='TE.start').reset_index(drop=True)

        # Liste pour stocker les indices des TE à conserver
        to_keep = []

        # Variable pour garder la fin du dernier TE conservé
        last_end = -float('inf')

        # Parcourir chaque ligne du DataFrame
        for i, row in group.iterrows():
            if row['TE.start'] - last_end > 500:
                to_keep.append(i)
                last_end = row['TE.end']

        # Filtrer le DataFrame pour conserver uniquement les TE sélectionnés
        filtered_group = group.loc[to_keep]
        filtered_dfs.append(filtered_group)
        
# Combiner les DataFrames filtrés pour chaque génome
filtered_te_df = pd.concat(filtered_dfs).reset_index(drop=True)

# Sauvegarder le DataFrame filtré dans un nouveau fichier CSV
filtered_te_df.to_csv('filtered_te_data.csv', index=False)

print(f'Original number of TEs: {sum([len(pd.read_csv(genome_file, delimiter="\t", skiprows=1)) for genome_file in genome_files])}')
print(f'Filtered number of TEs: {len(filtered_te_df)}')
```
